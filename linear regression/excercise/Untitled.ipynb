```python
import pandas as pd
import numpy as np
from sklearn import linear_model
import matplotlib.pyplot as plt
```

```python
df_cap= pd.read_csv('canada_per_capita_income.csv')
df_cap
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (us$)")
plt.scatter(df_cap['year'], df_cap['per capita income (US$)'], color='red', marker='*')

```

```python
model= linear_model.LinearRegression()
model.fit(df_cap[['year'], df_cap['per capita income (US$)']
```

```python
p=model.predict([[2020]])
p
```

```python
%matplotlib inline
plt.xlabel("year")
plt.ylabel("per capita income (us$)")
plt.scatter(df_cap['year'], df_cap['per capita income (US$)'], color='red', marker='*')
plt.plot(df_cap.year, model.predict(df_cap[['year']]), color= 'blue')
```

```python
new_row = pd.DataFrame([{'year': 2020}])
df_cap = pd.concat([df_cap, new_row], ignore_index=True)

#✅ ignore_index=True tells pandas to reset the index after adding the row.
df_cap
```

```python
df_cap = df_cap.drop(index=48)
df_cap
```
